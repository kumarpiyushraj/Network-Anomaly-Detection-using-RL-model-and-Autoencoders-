```python
# import tensorflow as tf
# import torch
# import matplotlib.pyplot as plt
# import seaborn as sns
# from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc

print("Num GPUs Available (TensorFlow):", len(tf.config.list_physical_devices('GPU')))
print("PyTorch CUDA available:", torch.cuda.is_available())

# Configure TensorFlow GPU memory growth (if a GPU is available)
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)

# import numpy as np
# import pandas as pd
# import glob

# Loading the CSV files
all_files = glob.glob("C:\\Users\\dell\\Downloads\\Dataset\\*.csv")  # Adjust path
df_list = [pd.read_csv(f) for f in all_files]
df = pd.concat(df_list, axis=0, ignore_index=True)

# Basic Info on the Dataset
print("\n=== Dataset Shape ===")
print(df.shape)

print("\n=== First 5 Rows ===")
print(df.head())

print("\n=== Column Names ===")
print(df.columns)

# Checking label distribution
df.columns = df.columns.str.strip().str.lower()
df['label'] = df['label'].apply(lambda x: 0 if str(x).lower() == 'benign' else 1)

print("\n=== Label Distribution ===")
print(df['label'].value_counts())

# Quick Outlier Check on numeric columns
# We'll do a simple IQR-based outlier count for demonstration
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
outlier_count = 0

for col in numeric_cols:
    if col == 'label':
        continue
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    # define outlier as outside [Q1-1.5*IQR, Q3+1.5*IQR]
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    outlier_count += col_outliers

print(f"\n=== Approx Outlier Count (IQR-based, across numeric cols) ===")
print(outlier_count, "outliers found (may include duplicates across columns).")

# Dropping columns not needed & handling NaN/Inf
cols_to_drop = ['destination port', 'flow duration']
df = df.drop([c for c in cols_to_drop if c in df.columns], axis=1)

df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(axis=1, thresh=int(0.7 * len(df)))  # drop columns with >30% missing
df = df.fillna(df.median())


# Splitting & Scaling Data
X = df.drop('label', axis=1).values
y = df['label'].values

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Cleaning any remaining NaN/Inf in training data
X_train = np.nan_to_num(X_train, nan=0.0, posinf=1e5, neginf=-1e5)

print("\n=== Final Train/Test Shapes ===")
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_test:", X_test.shape, "y_test:", y_test.shape)


# STEP 4: Autoencoder Training
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

input_dim = X_train.shape[1]
latent_dim = 32

# Autoencoder architecture
input_layer = Input(shape=(input_dim,))
encoded = Dense(128, activation='relu', kernel_regularizer=l2(1e-5))(input_layer)
encoded = Dropout(0.3)(encoded)
encoded = Dense(64, activation='relu', kernel_regularizer=l2(1e-5))(encoded)
encoded = Dropout(0.3)(encoded)
encoded = Dense(latent_dim, activation='relu')(encoded)

decoded = Dense(64, activation='relu')(encoded)
decoded = Dropout(0.3)(decoded)
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=1e-4, clipvalue=1.0), loss='mse')

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
]

history = autoencoder.fit(
    X_train, X_train,
    epochs=50,
    batch_size=2048,
    validation_split=0.1,
    callbacks=callbacks,
    verbose=1
)

# Plotting Autoencoder Loss Graph
plt.figure(figsize=(8, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Autoencoder Loss')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()

# Extracting encoder
encoder = Model(inputs=input_layer, outputs=encoded)


# RL Environment with Balanced Reward
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.env_util import make_vec_env

class AnomalyDetectionEnv(gym.Env):
    """
    Balanced reward shaping:
      +2 for correct malicious
      +2 for correct benign
      -3 for missed malicious
      -3 for false positive
    This tries to push for both high recall (don't miss malicious)
    and high precision (avoid false positives).
    """
    def __init__(self, X, y, encoder, latent_dim, max_steps=3000):
        super().__init__()
        self.X = X.astype(np.float32)
        self.y = y
        self.encoder = encoder
        self.latent_dim = latent_dim
        self.max_steps = max_steps
        self.current_idx = 0
        self.step_count = 0

        self.action_space = spaces.Discrete(2)  # 0=benign, 1=malicious
        self.observation_space = spaces.Box(
            low=-np.inf, high=np.inf, shape=(latent_dim,), dtype=np.float32
        )

    def reset(self, seed=None, options=None):
        self.current_idx = 0
        self.step_count = 0
        return self._get_encoded_state(0), {}

    def step(self, action):
        true_label = self.y[self.current_idx]

        if action == true_label:
            # correct classification
            reward = 2
        else:
            # incorrect classification
            reward = -3

        self.current_idx = (self.current_idx + 1) % len(self.X)
        self.step_count += 1

        done = self.step_count >= self.max_steps
        next_state = (
            self._get_encoded_state(self.current_idx) if not done else np.zeros(self.latent_dim)
        )

        return next_state, reward, done, False, {}

    def _get_encoded_state(self, idx):
        return self.encoder(self.X[idx:idx+1]).numpy()[0]

env = AnomalyDetectionEnv(X_train, y_train, encoder, latent_dim, max_steps=3000)
env = Monitor(env)
env = make_vec_env(lambda: env, n_envs=1)


# RL Agent Training (PPO)
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import EvalCallback

eval_env = AnomalyDetectionEnv(X_train, y_train, encoder, latent_dim, max_steps=3000)
eval_env = Monitor(eval_env)

eval_callback = EvalCallback(
    eval_env,
    best_model_save_path="./best_model/",
    eval_freq=4000,
    deterministic=True
)

model = PPO(
    "MlpPolicy",
    env,
    policy_kwargs={"net_arch": {"pi": [128, 128], "vf": [128, 128]}},
    learning_rate=1e-3,
    n_steps=1024,
    batch_size=512,
    gamma=0.995,
    clip_range=0.2,
    ent_coef=0.04,
    verbose=1,
    tensorboard_log="./ppo_logs/",
    device="auto"
)

model.learn(
    total_timesteps=120000,
    callback=eval_callback,
    progress_bar=True
)

model.save("ppo_anomaly_detector")


# Model Evaluation
from stable_baselines3.common.evaluation import evaluate_policy

test_env = AnomalyDetectionEnv(X_test, y_test, encoder, latent_dim, max_steps=3000)
test_env = Monitor(test_env)

mean_test_reward, std_test_reward = evaluate_policy(model, test_env, n_eval_episodes=10, deterministic=True)
print(f"\nTest Mean Reward: {mean_test_reward:.2f} +/- {std_test_reward:.2f}")

# Argmax Policy Classification
y_pred_argmax = []
obs, _ = test_env.reset()
for _ in range(len(X_test)):
    action, _ = model.predict(obs, deterministic=True)
    y_pred_argmax.append(action)
    obs, _, done, _, _ = test_env.step(action)
    if done:
        break

print("\n=== Classification Report (Argmax) ===")
print(classification_report(y_test[:len(y_pred_argmax)], y_pred_argmax))
cm_argmax = confusion_matrix(y_test[:len(y_pred_argmax)], y_pred_argmax)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_argmax, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Argmax Policy)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

precision_argmax, recall_argmax, _ = precision_recall_curve(y_test[:len(y_pred_argmax)], y_pred_argmax)
pr_auc_argmax = auc(recall_argmax, precision_argmax)

plt.figure(figsize=(8, 5))
plt.plot(recall_argmax, precision_argmax, label=f'PR AUC = {pr_auc_argmax:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (Argmax Policy)')
plt.legend()
plt.show()

# Custom Threshold Classification
def predict_with_threshold(model, obs, threshold=0.3):
    obs_tensor = torch.tensor(obs, dtype=torch.float32).to(model.device)
    with torch.no_grad():
        # For discrete actions, get_distribution gives a CategoricalDistribution
        distribution = model.policy.get_distribution(obs_tensor)
        probs = distribution.distribution.probs
    predicted_actions = (probs[:, 1] > threshold).long().cpu().numpy()
    return predicted_actions

threshold = 0.3
y_pred_threshold = []
obs, _ = test_env.reset()
for i in range(len(X_test)):
    action_array = predict_with_threshold(model, obs[None, :], threshold=threshold)
    action = action_array[0]
    y_pred_threshold.append(action)
    obs, _, done, _, _ = test_env.step(action)
    if done:
        break

print(f"\n=== Classification Report (Threshold={threshold}) ===")
print(classification_report(y_test[:len(y_pred_threshold)], y_pred_threshold))
cm_threshold = confusion_matrix(y_test[:len(y_pred_threshold)], y_pred_threshold)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_threshold, annot=True, fmt='d', cmap='Blues')
plt.title(f'Confusion Matrix (Custom Threshold={threshold})')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

precision_thr, recall_thr, _ = precision_recall_curve(y_test[:len(y_pred_threshold)], y_pred_threshold)
pr_auc_thr = auc(recall_thr, precision_thr)

plt.figure(figsize=(8, 5))
plt.plot(recall_thr, precision_thr, label=f'PR AUC = {pr_auc_thr:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (Threshold={threshold})')
plt.legend()
plt.show()

print("\n=== Program Complete ===")
```

```python

all_files = glob.glob("C:\\Users\\dell\\Downloads\\Dataset\\*.csv")
df_list = [pd.read_csv(f) for f in all_files]
df = pd.concat(df_list, axis=0, ignore_index=True)

# Cleaning column names and converting label to binary (0=Benign, 1=Malicious)
df.columns = df.columns.str.strip().str.lower()
df['label'] = df['label'].apply(lambda x: 0 if str(x).lower()=='benign' else 1)

# Dropping unnecessary columns
cols_to_drop = ['destination port', 'flow duration']
df = df.drop([c for c in cols_to_drop if c in df.columns], axis=1)

# Replacing infinities, dropping columns with >30% missing values, filling remaining NaNs with median
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(axis=1, thresh=int(0.7 * len(df)))
df = df.fillna(df.median())

print("\n=== Updated Dataset Metrics ===")
print(f"Total Records: {df.shape[0]}")
print(f"Total Features: {df.shape[1]}")
print("Class Distribution:")
print(df['label'].value_counts())

# Scaling and splitting data
X = df.drop('label', axis=1).values
y = df['label'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_temp, y_train, y_temp = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1, random_state=42, stratify=y_temp)

# For convolution, reshape each sample to (features, 1)
X_train_conv = X_train.reshape(-1, X_train.shape[1], 1)
X_val_conv = X_val.reshape(-1, X_val.shape[1], 1)
X_test_conv = X_temp.reshape(-1, X_train.shape[1], 1)


# Convolutional Autoencoder Implementation
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, UpSampling1D, Flatten, Dense, Reshape

input_shape = (X_train.shape[1], 1)
latent_dim = 32

# Building the Convolutional Autoencoder (ConvAE)
inputs = Input(shape=input_shape)
# Encoder
x = Conv1D(32, 3, activation='relu', padding='same')(inputs)
x = MaxPooling1D(2, padding='same')(x)
x = Conv1D(16, 3, activation='relu', padding='same')(x)
x = MaxPooling1D(2, padding='same')(x)
shape_before_flatten = tf.keras.backend.int_shape(x)  # (None, new_length, 16)
x = Flatten()(x)
latent = Dense(latent_dim, activation='relu')(x)

# Decoder
x = Dense(np.prod(shape_before_flatten[1:]), activation='relu')(latent)
x = Reshape(shape_before_flatten[1:])(x)
x = UpSampling1D(2)(x)
x = Conv1D(16, 3, activation='relu', padding='same')(x)
x = UpSampling1D(2)(x)
decoded = Conv1D(1, 3, activation='sigmoid', padding='same')(x)

conv_autoencoder = Model(inputs, decoded)
conv_autoencoder.compile(optimizer=Adam(learning_rate=1e-4, clipvalue=1.0), loss='mse')

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
]

history = conv_autoencoder.fit(
    X_train_conv, X_train_conv,
    epochs=50,
    batch_size=4096,  
    validation_data=(X_val_conv, X_val_conv),
    callbacks=callbacks,
    verbose=1
)

plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Convolutional Autoencoder Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()

# Extract encoder: flatten the latent vector
encoder_output = Model(inputs, latent)


# RL Environment Setup using Encoder Output
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.env_util import make_vec_env

class AnomalyDetectionEnv(gym.Env):
    def __init__(self, X, y, encoder, latent_dim, max_steps=3000):
        super().__init__()
        self.X = X.astype(np.float32)
        self.y = y
        self.encoder = encoder
        self.latent_dim = latent_dim
        self.max_steps = max_steps
        self.current_idx = 0
        self.step_count = 0
        self.action_space = spaces.Discrete(2)  # 0 = benign, 1 = malicious
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(latent_dim,), dtype=np.float32)
    
    def reset(self, seed=None, options=None):
        self.current_idx = 0
        self.step_count = 0
        return self._get_encoded_state(0), {}
    
    def step(self, action):
        true_label = self.y[self.current_idx]
        reward = 2 if action == true_label else (-4 if true_label == 1 else -1)
        self.current_idx = (self.current_idx + 1) % len(self.X)
        self.step_count += 1
        done = self.step_count >= self.max_steps
        next_state = self._get_encoded_state(self.current_idx) if not done else np.zeros(self.latent_dim)
        return next_state, reward, done, False, {}
    
    def _get_encoded_state(self, idx):
        # For each sample, reshape it to (features, 1) for encoder input
        sample = self.X[idx:idx+1].reshape(-1, self.X.shape[1], 1)
        return self.encoder(sample).numpy()[0]

env = AnomalyDetectionEnv(X_train, y_train, encoder_output, latent_dim, max_steps=3000)
env = Monitor(env)
env = make_vec_env(lambda: env, n_envs=1)

# RL Agent Training (PPO)
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import EvalCallback

eval_env = AnomalyDetectionEnv(X_train, y_train, encoder_output, latent_dim, max_steps=3000)
eval_env = Monitor(eval_env)

eval_callback = EvalCallback(
    eval_env,
    best_model_save_path="./best_model/",
    eval_freq=4000,
    deterministic=True
)

model = PPO(
    "MlpPolicy",
    env,
    policy_kwargs={"net_arch": {"pi": [128, 128], "vf": [128, 128]}},
    learning_rate=5e-4,
    n_steps=512,
    batch_size=256,
    gamma=0.995,
    gae_lambda=0.95,
    vf_coef=1.0,
    clip_range=0.2,
    ent_coef=0.02,
    verbose=1,
    tensorboard_log="./ppo_logs/",
    device="auto"
)

model.learn(
    total_timesteps=120000,
    callback=eval_callback,
    progress_bar=True
)

model.save("ppo_anomaly_detector")


# Model Evaluation on Test Data
from stable_baselines3.common.evaluation import evaluate_policy
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc

test_env = AnomalyDetectionEnv(X_test, y_test, encoder_output, latent_dim, max_steps=3000)
test_env = Monitor(test_env)
mean_test_reward, std_test_reward = evaluate_policy(model, test_env, n_eval_episodes=10, deterministic=True)
print(f"\nTest Mean Reward: {mean_test_reward:.2f} +/- {std_test_reward:.2f}")

# Argmax Policy Classification
y_pred_argmax = []
obs, _ = test_env.reset()
for _ in range(len(X_test)):
    action, _ = model.predict(obs, deterministic=True)
    y_pred_argmax.append(action)
    obs, _, done, _, _ = test_env.step(action)
    if done:
        break

print("\n=== Classification Report (Argmax) ===")
print(classification_report(y_test[:len(y_pred_argmax)], y_pred_argmax))
cm_argmax = confusion_matrix(y_test[:len(y_pred_argmax)], y_pred_argmax)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_argmax, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Argmax Policy)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

precision_argmax, recall_argmax, _ = precision_recall_curve(y_test[:len(y_pred_argmax)], y_pred_argmax)
pr_auc_argmax = auc(recall_argmax, precision_argmax)

plt.figure(figsize=(8, 5))
plt.plot(recall_argmax, precision_argmax, label=f'PR AUC = {pr_auc_argmax:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (Argmax Policy)')
plt.legend()
plt.show()

# Custom Threshold Classification
import torch

def predict_with_threshold(model, obs, threshold=0.3):
    obs_tensor = torch.tensor(obs, dtype=torch.float32).to(model.device)
    with torch.no_grad():
        distribution = model.policy.get_distribution(obs_tensor)
        probs = distribution.distribution.probs  # shape [batch_size, 2]
    predicted_actions = (probs[:, 1] > threshold).long().cpu().numpy()
    return predicted_actions

for threshold in [0.25, 0.3, 0.35]:
    y_pred_threshold = []
    obs, _ = test_env.reset()
    for _ in range(len(X_test)):
        action_array = predict_with_threshold(model, obs[None, :], threshold=threshold)
        action = action_array[0]
        y_pred_threshold.append(action)
        obs, _, done, _, _ = test_env.step(action)
        if done:
            break

    print(f"\n=== Classification Report (Threshold={threshold}) ===")
    print(classification_report(y_test[:len(y_pred_threshold)], y_pred_threshold))
    cm_threshold = confusion_matrix(y_test[:len(y_pred_threshold)], y_pred_threshold)

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm_threshold, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix (Custom Threshold={threshold})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    precision_thr, recall_thr, _ = precision_recall_curve(y_test[:len(y_pred_threshold)], y_pred_threshold)
    pr_auc_thr = auc(recall_thr, precision_thr)

    plt.figure(figsize=(8, 5))
    plt.plot(recall_thr, precision_thr, label=f'PR AUC = {pr_auc_thr:.2f}')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall Curve (Threshold={threshold})')
    plt.legend()
    plt.show()

print("\n=== Program Complete ===")
```
